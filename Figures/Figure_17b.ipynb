```python
# selecte ss316 steels
tensile_df_plots = tensile_df_plots[tensile_df_plots['Material'].isin(['JPCA', 'SS316', 'SS316-A', 'SS316-B', 'SS316-C'])]
```

```python
# implment spline fitting for ss316 steels, show impact of thickness-to-grain-size (T/GS) ratio on TE
from sklearn.utils import resample
# Assuming df_tgs is already defined as per your code
df_tgs = tensile_df_plots[['Reference', 'Grain size (um)', 'Thickness (mm)', 'Total Elongation (%)', 'Treatment – Three Categories']].dropna()
df_tgs['T/GS'] = df_tgs['Thickness (mm)']*1000/df_tgs['Grain size (um)']
df_tgs = df_tgs[df_tgs['Reference'] ==21]


index_ids_to_remove = [922, 923, 925, 970]
df_tgs = df_tgs.drop(index_ids_to_remove)

#Extract the series
x = df_tgs['T/GS']
y = df_tgs['Total Elongation (%)']


x = x.to_numpy()[:, np.newaxis]
y = y.to_numpy()



knots = np.array([0.01, 6.21, 25]).reshape(-1, 1)

# Create and fit the spline model
spline_transformer = SplineTransformer(degree=2, knots=knots)
model = make_pipeline(spline_transformer, LinearRegression())
model.fit(x,y)

# Create spline for plotting
x_spline = np.linspace(x.min(), x.max(), 1000)
x_spline = x_spline[:, np.newaxis]
y_spline = model.predict(x_spline)





# Number of bootstrap samples
n_bootstrap = 3000
bootstrapped_fits = []

# Bootstrap resampling loop
for _ in range(n_bootstrap):
    # Resample the data
    x_resampled, y_resampled = resample(x, y)
    
    # Fit the pipeline to the resampled data
    model.fit(x_resampled, y_resampled)
    
    # Evaluate the model on the original x values and store the result
    y_resampled_fit = model.predict(x_spline)
    bootstrapped_fits.append(y_resampled_fit)

# Convert the list of bootstrap fits to a NumPy array for easier manipulation
fits_array = np.array(bootstrapped_fits)

# Calculate the mean and confidence intervals
fit_mean = np.mean(fits_array, axis=0)
fit_lower = np.percentile(fits_array, 2.5, axis=0)
fit_upper = np.percentile(fits_array, 97.5, axis=0)

# Plot the original data and the spline fit
plt.plot(x, y, 'o', markersize=5)
plt.plot(x_spline-0.5, y_spline, 'r-')

# Plot the confidence interval
plt.fill_between(x_spline.ravel()-0.5, fit_lower, fit_upper, color='gray', alpha=0.5)

plt.xlabel('Thickness-to-Grain Size (T/GS) Ratio', fontsize=16)
plt.ylabel('Total Elongation (%)', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.axvline(x=15, color='g', linestyle='--', linewidth=2)
# Add grid to the plot
plt.xlim(left=0)
plt.ylim(0,100)
plt.grid(True) 
plt.show()

```
